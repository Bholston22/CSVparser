```python
Filename = 'transcript.csv'
```

```python
df = pd.read_csv(Filename) #This is where I read the transcript from excel
```

```python
df.dtypes # Here I wanted to make sure the column grades was an int so I can use sum()
```

```python
def overallGrade(Filename):
    data = [] # all the grades put into a list
    courses = 0 # need the number of courses to get gpa
    with open (Filename, 'r') as handle:
        filereader = csv.DictReader(handle) 
        for grades in filereader:
            grades = int(grades['Grade'])
            if grades > 0:
                data.append(grades)
                courses += 1
            else:
                pass
    overallgrades = sum(data)/len(data)

    print("Overall Grades:", overallgrades)
    print("GPA:", overallgrades /courses)
    print("Grades:" )
    return data
    
```
